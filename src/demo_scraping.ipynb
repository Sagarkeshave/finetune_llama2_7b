```python
import pandas as pd
from bs4 import BeautifulSoup
import requests, time
```

```python
# Function to scrape with retry mechanism
def scrape_with_retry(url, max_retries=3, retry_delay=5):
    for attempt in range(max_retries):
        try:
            headers = {
                'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.111 Safari/537.36'
            }
            response = requests.get(url, headers= headers, timeout=5)
            response.raise_for_status()  # Raise an exception if response status is not 2xx
            return response


        except requests.exceptions.RequestException as e:
            print(f"Attempt {attempt + 1} failed: {e}")
            if attempt < max_retries - 1:
                print(f"Retrying in {retry_delay} seconds...")
                time.sleep(2)
            else:
                print("Max retries exceeded.")
                raise
```

```python
def scraper(url, df):

      response = scrape_with_retry(url)

      soup = BeautifulSoup(response.content, "html.parser")

      dr_names = soup.find_all("h2", class_ = "elementor-heading-title elementor-size-default")

      name_list = []
      for i in dr_names:
        name_list.append(i.text)

      # print(name_list)
      df["Dr_name"] = name_list

      doc_profile = soup.find_all( "div" , { "data-widget_type" : "text-editor.default" })

      doc_profile_list = []
      for i in doc_profile:
        doc_profile_list.append(i.text)

      df["doc_profile_"] = doc_profile_list

      df = pd.DataFrame(df)

      return df
```
